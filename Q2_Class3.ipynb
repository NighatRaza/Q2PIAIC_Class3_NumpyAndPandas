```python
# in  or merge operator should be used over list datatype and not object datatype
ubl_database = np.arange(1,200)
sbp = [20,2,26,7,3]

display(ubl_database)
display(sbp)
```

```python
sbp in ubl_database
```

```python
# method 1 by for loop/ slower way
for i in sbp:
    print(i, i in ubl_database)
```

```python
l = []
for i in sbp:
    if i in ubl_database:
        print(i)
        l.append(i)
display(l)
```

```python
ubl_database = np.arange(1,30)
sbp = [20,2,26,7,3,1000,5000,3,25000]

display(ubl_database)
display(sbp)
```

```python
method 2 bydefault function/ faster way
# checks each value in list and return true/false
np.in1d(sbp,ubl_database)
```

```python
sbp = np.array(sbp)
print(type(sbp))
sbp
```

```python
sbp[np.in1d(sbp,ubl_database)]
```

```python
# not operator / return remaining values
display(~np.in1d(sbp,ubl_database))
```

```python
sbp[~np.in1d(sbp,ubl_database)]
```
